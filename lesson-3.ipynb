```python
import pandas as pd
d = pd.read_excel('kc_house_data.xlsx')
 
df = pd.DataFrame(columns= d.columns[0].split(','))
for j,k in enumerate(d.columns[0].split(',')):
    df[k] = [i.split(',')[j] for i in d[d.columns[0]]]
    

df['sqft_lot'] = pd.to_numeric(df['sqft_lot'], errors='coerce') 
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
 
print('Стоимость кв.м в первых 10-ти домах:\n', df['price_per_sq_lot'].head(10))

df = df.drop(['date', 'zipcode', 'lat', 'long'], axis = 1)
df
 

```
